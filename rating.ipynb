```python
import pandas as pd 
from textblob import TextBlob
```

```python
data = pd.read_csv("chrome_reviews.csv")
data.head()
```

```python
data.shape
```

```python
# new df for id,text,star 
df = data[['ID','Text','Star']]
df.shape
```

```python
# dropping high rating rows
df = df[df.Star != 5]
df = df[df.Star != 4]
df = df[df.Star != 3]
```

```python
df
```

```python
df.Star.value_counts()
```

```python
# use textblob for sentiment category.
list_sentiment = []
for i in df["Text"]:
    score = TextBlob(i).sentiment[0]
    if (score > 0):
        list_sentiment.append('Positive')
    elif (score < 0):
        list_sentiment.append('Negative')
    else:
        list_sentiment.append('Neutral')
```

```python
df["Sentiment"]=list_sentiment
df.head()
```

```python
df.shape
```

```python
df = df[df.Sentiment == 'Positive']
df.shape
```

```python
df
```

```python
df.to_csv("positive_rating.csv", index=False)
```
